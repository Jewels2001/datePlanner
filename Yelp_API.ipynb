```python
import requests
from tabulate import tabulate

def search_restaurants(api_key, location):
    url = "https://api.yelp.com/v3/businesses/search"
    headers = {
        "Authorization": f"Bearer {api_key}"
    }
    params = {
        "term": "restaurants",
        "location": location
        #"limit": 10
    }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        restaurant_info = []
        for business in data["businesses"]:
            name = business["name"]
            rating = business.get("rating", "Rating not available")
            price = business.get("price", "Price not available")
            restaurant_info.append([name, rating, price])
        print(tabulate(restaurant_info, headers=["Name", "Rating", "Price"], tablefmt="grid"))
    else:
        print("Error:", response.status_code)

if __name__ == "__main__":
    api_key = "VMAbzvQFvaL1Rh8st7-2ECtvKggFIZcja2LYG1WDJrKyhiGNIHno-wiCX32VmIS1OPzIC4KDeBWDAz4hqHnSHmNYM7QD-4CViREgX5V5b1R2YdI941PaSMM5eh3IZXYx"
    location = input("Enter a location to search for restaurants: ")
    search_restaurants(api_key, location)
```

```python
import requests
from tabulate import tabulate

def search_places(api_key, location, term):
    url = "https://api.yelp.com/v3/businesses/search"
    headers = {
        "Authorization": f"Bearer {api_key}"
    }
    params = {
        "term": term,
        "location": location
        #"limit": 10  # Adjust as needed
    }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        print(data)
        places_info = []
        for place in data["businesses"]:
            name = place["name"]
            rating = place.get("rating", "Rating not available")
            price = place.get("price", "Price not available")
            is_closed = place.get("is_closed", "N/A")
            address = ', '.join(place["location"]["display_address"])
            display_phone = place.get("display_phone", "Number not available")
            places_info.append([name, rating, price, is_closed, address, display_phone])
        return places_info
    else:
        print("Error:", response.status_code)
        return None

if __name__ == "__main__":
    api_key = "VMAbzvQFvaL1Rh8st7-2ECtvKggFIZcja2LYG1WDJrKyhiGNIHno-wiCX32VmIS1OPzIC4KDeBWDAz4hqHnSHmNYM7QD-4CViREgX5V5b1R2YdI941PaSMM5eh3IZXYx"
    location = input("Enter a location: ")

    # Search for restaurants
    restaurant_info = search_places(api_key, location, "restaurants")
    if restaurant_info:
        print("\nRestaurant Information:")
        print(tabulate(restaurant_info, headers=["Name", "Rating", "Price", "Open?", "Address", "Phone Number"], tablefmt="grid"))

    # Search for recreational centers
    recreational_info = search_places(api_key, location, "recreation")
    if recreational_info:
        print("\nRecreational Center Information:")
        print(tabulate(recreational_info, headers=["Name", "Rating", "Price", "Open?", "Address", "Phone Number"], tablefmt="grid"))
```
